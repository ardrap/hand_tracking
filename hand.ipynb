```python
converted_img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(converted_img)
```

```python
mp_hands=mp.solutions.hands  #hands--->fun
```

```python
hands=mp_hands.Hands(static_image_mode=True,max_num_hands=2,min_detection_confidence=0.3) #(imgvideo,max_number_hands,)
```

```python
mp_drawing=mp.solutions.drawing_utils #for dots in hand
```

```python
results=hands.process(converted_img)
```

```python
print(results.multi_handedness)   #accuracy of hands
```

```python
print(results.multi_hand_landmarks) #hand_dot positions
```

```python
if results.multi_hand_landmarks:
    for hand_no,hand_landmarks in enumerate(results.multi_hand_landmarks):
        mp_drawing.draw_landmarks(image=converted_img,landmark_list=hand_landmarks,connections=mp_hands.HAND_CONNECTIONS) #connections -->used to connect dots/points
    plt.imshow(converted_img)
```
